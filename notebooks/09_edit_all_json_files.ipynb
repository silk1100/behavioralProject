```python
EXPERIMENT_PATH = "../experiments"
SEVERITY_GROUP_DICT = {
    'mild':['mild'],
    'moderate':['moderate','mod', 'moder','moderat'],
    'sever':['sever','sev', 'severe','seve'],
    'TD':['td'],
}

TEMP_EXPERIMENT = {
    "data_repr": "perc",
    "normalizer": "minmax",
    "DD": {
        "srs_type": "cog",
        "severity_group": ["mild", "TD"],
        "age_group": None,
        "divide_data": False,
        "balance": False
    },
    "FS": {
        "est": ["lsvm", "rf", "xgb", "lr"],
        "cv": 5,
        "scoring": "balanced_accuracy",
        "n_jobs": -1,
        "verbose": 0,
        "step": 1,
        "min_features_to_select": 1
    },
    "ML": {
        "est": ["lr", "lsvm", "svm", "xgb", "rf", "ridge", "gnb","nn"],
        "cv": 5,
        "scoring": "balanced_accuracy",
        "n_jobs": -1,
        "verbose": 0,
        "hyper_search_type": "random",
        "n_iter": 1500}
}
```

```python
import json
import os
```

# Goals
1. Read experiments folder name
2. Extract age range
3. Update all internal json file to be within the same age range

## Updated Age

```python
for exp_fldr in [x for x in os.listdir(EXPERIMENT_PATH) if os.path.isdir(os.path.join(EXPERIMENT_PATH, x))]:
    fullpath = os.path.join(EXPERIMENT_PATH, exp_fldr)
    age_groups = exp_fldr.split('_')[0].split('t')
    endage = age_groups[-1]
    startage = age_groups[1].split('een')[1]
    
    try:
        s = int(startage)
    except Exception:
        startage = None
        
    try:
        s = int(endage)
    except Exception:
        endage = None
        
    if (startage is not None) and (endage is not None):
        age_group=[float(startage), float(endage)]
    elif (startage is not None) and (endage is None):
        age_group=[float(startage), None]
    else:
        age_group = None
        
    for jsnfile in [x for x in os.listdir(fullpath) if x.endswith('.json')]:
        fullfilepath = os.path.join(fullpath, jsnfile)
        with open(fullfilepath, 'r') as f:
            data = json.load(f)
        data["DD"]['age_group'] = age_group
        with open(fullfilepath, 'w') as f:
            json.dump(data, f)
        
```

## Updated severity groups

```python
def find_groupname_in_dict(keydict, group_1):        
    found = True
    group_1 = group_1.lower()
    if group_1 not in keydict:
        found=False
        for key, values in keydict.items():
            if group_1 in values:
                found = True
                group_1 = key
                break
    if found:
        return group_1
    raise ValueError(group_1," cant be found in ",keydict)

    
```

```python
for exp_fldr in [x for x in os.listdir(EXPERIMENT_PATH) if os.path.isdir(os.path.join(EXPERIMENT_PATH, x))]:
    fullexppath = os.path.join(EXPERIMENT_PATH, exp_fldr)
    group_1 = exp_fldr.split('_')[0]
    group_2 = exp_fldr.split('_')[1]

    group_1 = find_groupname_in_dict(SEVERITY_GROUP_DICT, group_1)
    group_2 = find_groupname_in_dict(SEVERITY_GROUP_DICT, group_2)
    
    for subexp in [x for x in os.listdir(fullexppath) if os.path.isdir(os.path.join(fullexppath, x))]:
        fullpath = os.path.join(fullexppath, subexp)
        print(fullpath)
        for jsnfile in [x for x in os.listdir(fullpath) if x.endswith('.json')]:
            fullfilepath = os.path.join(fullpath, jsnfile)
            with open(fullfilepath, 'r') as f:
                data = json.load(f)
            data["DD"]['severity_group'] = [group_1, group_2]
            with open(fullfilepath, 'w') as f:
                json.dump(data, f)
            print(f"wrote data in {fullfilepath}")
            
```
