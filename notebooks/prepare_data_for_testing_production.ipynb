```python
# data_dir = '../data/feature_extraction/medianMinusPlus/'
data_dir = '../data/feature_extraction/percentile/'
```

```python
df_raw = pd.read_csv(os.path.join(data_dir, 'raw.csv'), index_col=0)
```

```python
df_pheno = pd.read_csv(os.path.join(data_dir,'updated_pheno.csv'), index_col=0)
```

```python
df_meta = df_pheno[['subj_id','AGE_AT_SCAN ', 'SEX', 'SRS_TOTAL_T','SRS_AWARENESS_T',
                    'SRS_COGNITION_T','SRS_COMMUNICATION_T','SRS_MOTIVATION_T','SRS_MANNERISMS_T']]
```

```python
df_meta.set_index('subj_id', inplace=True)
```

```python
df = df_raw.drop(['SEX','AGE_AT_SCAN '], axis=1).join(df_meta)
```

```python
df.head()
```

```python
SRS_SCORES_MAP = {
    'TD': (0, 59),
    'mild':(60, 65),
    'moderate':(66, 75),
    'sever':(76, 1000)
}
```

```python
def add_severity_label_withnans(row, srs_type):
    if np.isnan(row[srs_type]):
        return 'sever' if row['DX_GROUP']==1 else 'TD'
    elif row[srs_type]<=59:
        return 'TD'
    elif row[srs_type] <= 65:
        return 'mild'
    elif row[srs_type] <= 75:
        return 'moderate'
    else:
        return 'sever'
```

```python
df['severity_label_comm'] = df.apply(lambda x: add_severity_label_withnans(x, 'SRS_COMMUNICATION_T'), axis=1)
df['severity_label_awa'] = df.apply(lambda x: add_severity_label_withnans(x, 'SRS_AWARENESS_T'), axis=1)
df['severity_label_cog'] = df.apply(lambda x: add_severity_label_withnans(x, 'SRS_COGNITION_T'), axis=1)
df['severity_label_mot'] = df.apply(lambda x: add_severity_label_withnans(x, 'SRS_MOTIVATION_T'), axis=1)
df['severity_label_tot'] = df.apply(lambda x: add_severity_label_withnans(x, 'SRS_TOTAL_T'), axis=1)
df['severity_label_man'] = df.apply(lambda x: add_severity_label_withnans(x, 'SRS_MANNERISMS_T'), axis=1)
```
